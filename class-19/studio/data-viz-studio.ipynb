```python
# Here is the setup of the dataframe. Feel free to use .info() or .head() to 
## get a better understanding of what is inside the dataframe!

import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib import style

df = pd.read_csv("books.csv", on_bad_lines='skip')

years = []

for y in df['publication_date']:
    date = y.split("/")
    years.append(date[2])

df["year"] = years

df.head()
```

With your dataframe set up, time to start creating visualizations!

```python
# Use your Python skills to create a visualization here.

plt.scatter(x = df['ratings_count']/1000000, y = df['text_reviews_count']/10000)
plt.xlabel('Ratings Count (in millions)')
plt.ylabel('Text Reviews Count (in tens of thousands)')
plt.title('Correlation between Ratings and Text Reviews')
```

```python
# Use your Python skills to create a visualization here.

publisher_names = df[df['publisher'].notnull()]['publisher'].unique()
sum_by_publisher = []

for p in publisher_names:
    sum = int(df[df['publisher'] == p]['bookID'].sum())
    sum_by_publisher.append(sum)

df_pub = pd.DataFrame()

df_pub['publisherID'] = publisher_names
df_pub['book_total'] = sum_by_publisher


plot = sb.barplot(df_pub, x='publisherID', y='book_total')

plt.xlabel('Publisher Name')
plt.ylabel('Total Number of Books Published')
plt.title('Number of Books Published by Publisher')


# plt.bar(x=publishers, height=books_by_publisher, width=0.5, color='purple')
```
