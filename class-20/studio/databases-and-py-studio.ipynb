## Working with the Data

Using Pandas or SQL, find the answer to these 2 questions:
1. How many of the total shows (full csv list) are on each streaming service?
2. What percentage of these total shows is available on each streaming service?

**Hint**:

Use the pandas `query` method to filter the data, and then the Python `len` method to find it's length. [Relevant Link](https://www.geeksforgeeks.org/ways-to-filter-pandas-dataframe-by-column-values/)

```python
# Code here

dataN = df.query('Netflix == 1')
dataH = df.query('Hulu == 1')
dataP = df.query('Prime == 1')
dataD = df.query('Disney == 1')

print(len(dataN))
print(len(dataH))
print(len(dataP))
print(len(dataD))

percN = (len(dataN)/len(df))*100
percH = (len(dataH)/len(df))*100
percP = (len(dataP)/len(df))*100
percD = (len(dataD)/len(df))*100

print(round(percN,1))
print(round(percH,1))
print(round(percP,1))
print(round(percD,1))
```


Now join your `watchlist` data to the `shows` data using pandas or SQL. Verify that you joined the data correctly.

Using this related dataset, come up with analytic code that answers these questions:
1. The number of watchlist shows each streaming service has
2. The percentage of your overall watchlist each streaming service has


```python
# Code here

df_join = pd.merge(df, wishlistdf, how = 'left', left_on = 'Title', right_on = 'title')

pd.set_option('display.max_columns', None)
# display(df_join)

dataN = df_join.query('Netflix == 1')
dataH = df_join.query('Hulu == 1')
dataP = df_join.query('Prime == 1')
dataD = df_join.query('Disney == 1')

watchlistN = dataN['importance_rank'].notna().sum()
watchlistH = dataH['importance_rank'].notna().sum()
watchlistP = dataP['importance_rank'].notna().sum()
watchlistD = dataD['importance_rank'].notna().sum()

print(watchlistN)
print(watchlistH)
print(watchlistP)
print(watchlistD)

watchlistpercN = (watchlistN/10)*100
watchlistpercH = (watchlistH/10)*100
watchlistpercP = (watchlistP/10)*100
watchlistpercD = (watchlistD/10)*100

print(watchlistpercN)
print(watchlistpercH)
print(watchlistpercP)
print(watchlistpercD)
```
